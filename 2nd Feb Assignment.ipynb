### Q1) What are the characteristics of the tuples? Is tuple immutable?
->Tuples are an ordered collection of elements, typically written as a comma-separated sequence of values enclosed in parentheses. The characteristics of tuples include:

Ordered: The elements in a tuple are ordered, which means that they have a specific position or index in the tuple.

Immutable: Tuples are immutable, which means that once a tuple is created, its contents cannot be changed. This is in contrast to lists, which are mutable and can be modified after creation.

Heterogeneous: Tuples can contain elements of different data types, including numbers, strings, and other tuples.

Hashable: Tuples are hashable, which means that they can be used as keys in a dictionary or as elements in a set.

### Q2) What are the two tuple methods in python? Give an example of each method. Give a reason why
### tuples have only two in-built methods as compared to Lists.
->There are two built-in methods for tuples in Python:
count: This method returns the number of times a specified element appears in a tuple. 
* example:
#### my_tuple = (1, 2, 3, 4, 4, 4)
#### count_of_4 = my_tuple.count(4)
#### print(count_of_4)  # Output: 3

index: This method returns the index of the first occurrence of a specified element in a tuple.
* example:
#### my_tuple = ('apple', 'banana', 'cherry', 'banana')
#### index_of_banana = my_tuple.index('banana')
#### print(index_of_banana)  # Output: 1


Tuples have only two built-in methods (count and index) compared to lists because tuples are immutable, and many of the list methods modify the list in place, which is not possible with tuples. Since tuples are not intended to be modified once created, they do not need as many methods as lists. Additionally, tuples are often used to represent fixed collections of data, and so the two built-in methods are usually sufficient for working with them.

Q4. Explain the difference between the union() and update() methods for a set. Give an example of
each method.

->Both union() and update() are methods used to combine two or more sets in Python, but they differ in their behavior.

The union() method returns a new set that contains all the elements from both sets without modifying the original sets.

Example:

* set1 = {1, 2, 3}
* set2 = {3, 4, 5}
* set3 = set1.union(set2)
* print(set3)  # Output: {1, 2, 3, 4, 5}

In the example above, set1 and set2 are combined using the union() method to create a new set set3 that contains all the unique elements from both sets.

The update() method, on the other hand, modifies the original set by adding all the elements from another set to it.

Example:

* set1 = {1, 2, 3}
* set2 = {3, 4, 5}
* set1.update(set2)
* print(set1)  # Output: {1, 2, 3, 4, 5}

In the example above, set2 is added to set1 using the update() method, and set1 is modified to contain all the elements from both sets.

In summary, the union() method returns a new set that contains all the elements from both sets without modifying the original sets, while the update() method modifies the original set by adding all the elements from another set to it.


Q5) What is a dictionary? Give an example. Also, state whether a dictionary is ordered or unordered.

->In Python, a dictionary is a collection of key-value pairs, where each key is associated with a value. Dictionaries are implemented as hash tables, which provide fast lookups and insertions, making them a very useful data structure.

Here's an example of a dictionary that maps the names of fruits to their corresponding colors:

* fruits = {'apple': 'red', 'banana': 'yellow', 'kiwi': 'green', 'orange': 'orange'}

In this dictionary, the keys are the names of the fruits ('apple', 'banana', etc.), and the values are their corresponding colors ('red', 'yellow', etc.). We can access the value associated with a key by using square brackets and the key as the index, like this:

* print(fruits['banana'])  # Output: 'yellow'

* Dictionaries are unordered, which means that the order of the keys may not be the same as the order in which they were added to the dictionary. However, starting from Python 3.7, dictionaries are guaranteed to maintain the insertion order. So, in modern versions of Python, the order in which keys are added to a dictionary is maintained. Nonetheless, it is important to note that the primary functionality of a dictionary is not based on its order. Instead, the keys are used to quickly and efficiently look up values.

Output:

{'language': 'Python', 'course': 'Data Science Masters', 'topics': ['Python', 'Machine Learning', 'Deep Learning']}

In the code above, we used the setdefault() method to add a new key-value pair to the dict1 dictionary. The first argument of setdefault() is the key that we want to add, and the second argument is the default value if the key does not already exist in the dictionary. If the key already exists, setdefault() returns its value.

Since the topics key does not exist in dict1, setdefault() adds it with the value ['Python', 'Machine Learning', 'Deep Learning']. If we print dict1, we can see that the topics key-value pair has been added to the dictionary.

Q8. What are the three view objects in dictionaries? Use the three in-built methods in python to display
these three view objects for the given dictionary.

dict1 = {'Sport': 'Cricket' , 'Teams': ['India', 'Australia', 'England', 'South Africa', 'Sri Lanka', 'New Zealand']}

->In Python, dictionaries have three built-in view objects:

* dict_keys: It is a view object that contains the keys of the dictionary.
* dict_values: It is a view object that contains the values of the dictionary.
* dict_items: It is a view object that contains the key-value pairs of the dictionary as tuples.

Here's an example of using these view objects with the dict1 dictionary:

#Creating a dictionary

dict1 = {'Sport': 'Cricket' , 'Teams': ['India', 'Australia', 'England', 'South Africa', 'Sri Lanka', 'New Zealand']}

#Using the view objects

keys_view = dict1.keys()

values_view = dict1.values()

items_view = dict1.items()

#Displaying the view objects

print("Keys view object:", keys_view)

print("Values view object:", values_view)

print("Items view object:", items_view)
